```python
import time
import requests
import ipywidgets as widgets
from IPython.display import display
import json
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
from matplotlib.backends.backend_pdf import PdfPages

style.use('fivethirtyeight')

def getUnix(_date, fromPicker = False):
    if fromPicker:
        day = _date.value
    else:
        day = _date.date()
    return int(time.mktime(day.timetuple()))

def forecast(location,unixTime):
    myurl = 'https://api.darksky.net/forecast/a5697e4c206dd1fe1f410c5c6fa7cd8d/{},{}?exclude=currently,minutely,hourly,alerts,flags'.format(location['coords'],unixTime)
    myPage = requests.get(myurl)
    weather = json.loads(myPage.content)

    stuff = weather['daily']['data'][0]
    print(location['name'])
    print("\t",stuff['summary'])
    print("\t",stuff['apparentTemperatureMin'])
    print("\t",stuff['apparentTemperatureMax'])
    try:
        print('\t {:.2f} inches'.format(stuff['precipIntensity']*24))
        print("\t",int(stuff['precipProbability']*100),"% chance")
    except:
        print("\tno rain")
    
    return stuff

def threeDaySummary(location):
    timeList = np.array([])
    tempList = np.array([])
    precipList = np.array([])
    summaryList = np.array([])
    cloudList = np.array([])
    rainList = []

    friday = 4
    now = datetime.datetime.now()
    untilFriday = friday-now.weekday()
    friDate = now + datetime.timedelta(days=untilFriday)
    satDate = now + datetime.timedelta(days=untilFriday+1)
    sunDate = now + datetime.timedelta(days=untilFriday+2)

    for day in [friDate,satDate,sunDate]:
        unixTime = getUnix(day)
        myurl = 'https://api.darksky.net/forecast/a5697e4c206dd1fe1f410c5c6fa7cd8d/{},{}?exclude=currently,minutely,alerts,flags'.format(location['coords'],unixTime)
        myPage = requests.get(myurl)
        weather = json.loads(myPage.content)

        for hr in weather['hourly']['data']:
            timeList = np.append(timeList,datetime.datetime.fromtimestamp(hr['time']))
            tempList = np.append(tempList,hr['temperature'])
            precipList = np.append(precipList,hr['precipIntensity'])
            cloudList = np.append(cloudList,int(hr['cloudCover']*100))
        myDay = weather['daily']['data'][0]
        summaryList = np.append(summaryList,myDay['summary'])
        try:
            prob,acuumalted = int(myDay['precipProbability']*100),myDay['precipIntensity']*24
        except:
            prob,acuumalted = [0,0]
        rainList.append((prob,round(acuumalted,2)))
        
    return [timeList,tempList,precipList,summaryList,cloudList,rainList,location['name']]

def plot3Day(inputs):
    timeList,tempList,precipList,summaryList,cloudList,rainList,title = inputs
    getHour = lambda dtme: dtme.strftime('%I\n%p').lower()
    convertToHourString = np.vectorize(getHour)
    tickResolution = 6
    theFigure = plt.figure()
    theFigure.set_size_inches(22,12)
    numRows,numColumns = 2,1
    
    precipPlot = plt.subplot2grid((numRows,numColumns), (0, 0),rowspan=1, colspan=1)
    precipPlot.plot(timeList,precipList,'-o',linewidth=.5)
    precipPlot.set_ylabel('Rain Intensity (inches/hr)',color='C0')
    precipPlot.set_ylim(-.01,.35)

    cloudPlot = precipPlot.twinx()
    cloudPlot.plot(timeList,cloudList,'-',color='C1',linewidth=.5)
    cloudPlot.set_ylabel('Cloud Cover (%)',color='C1',rotation=270, va='bottom')
    cloudPlot.grid(None)
    cloudPlot.set_xticks(timeList[::24])
    cloudPlot.set_xticks(timeList[::tickResolution],minor=True)
    cloudPlot.set_xticklabels([])
    cloudPlot.set_xticklabels(convertToHourString(timeList[::tickResolution]),minor=True)
    cloudPlot.set_ylim(-5,105)

    for i,dayName in enumerate(['Friday','Saturday','Sunday']):
        cloudPlot.annotate('{}\n{}'.format(dayName,summaryList[0]), xy=((i*2+1)/6, 1.15),xycoords='axes fraction',ha='center',fontsize='20')
        cloudPlot.annotate('{}'.format(rainList[i][1]), xy=((i*2+1)/6, 1.05),xycoords='axes fraction',ha='center',va='center',
                           fontsize=rainList[i][1]/2.5*30+15,color='C0',alpha=rainList[i][0]/100)
    
    
    tempPlot = plt.subplot2grid((numRows,numColumns), (1, 0),rowspan=1, colspan=1)
    tempPlot.plot(timeList,tempList,'-o',color='C2',linewidth=.5)
    tempPlot.set_xticks(timeList[::24])
    tempPlot.set_xticks(timeList[::tickResolution],minor=True)
    tempPlot.set_xticklabels([])
    tempPlot.set_xticklabels(convertToHourString(timeList[::tickResolution]),minor=True)
    tempPlot.set_ylabel('Temperature (F)',color='C2')
    tempPlot.set_ylim(45,105)
    tempPlot.set_title(title)

#     plt.show()
#     plt.savefig('{}.pdf'.format(title),bbox_inches='tight')
    return theFigure
    
def getWeekendForecast(b):
    with PdfPages('Climbing Weather.pdf') as pdf:
        NRG = {'coords':'38.072, -81.081','name':'New River Gorge'}
        rslts = threeDaySummary(NRG)
        pdf.savefig(plot3Day(rslts),bbox_inches='tight')

        SH = {'coords':'39.935, -76.385','name':'Safe Harbor'}
        rslts = threeDaySummary(SH)
        pdf.savefig(plot3Day(rslts),bbox_inches='tight')

        GF = {'coords':'38.992, -77.249','name':'Great Falls'}
        rslts = threeDaySummary(GF)
        pdf.savefig(plot3Day(rslts),bbox_inches='tight')

        SNCA = {'coords':'38.834, -79.366','name':'Seneca Rocks'}
        rslts = threeDaySummary(SNCA)
        pdf.savefig(plot3Day(rslts),bbox_inches='tight')

        AR = {'coords':'39.558, -77.599','name':'Annapolis Rock'}
        rslts = threeDaySummary(AR)
        pdf.savefig(plot3Day(rslts),bbox_inches='tight')
    
    

def showPickedForecast(b):
    print('{}/{}'.format(myDate.value.month,myDate.value.day))
    forecast(getUnix(myDate,fromPicker=True))
    
myDate = widgets.DatePicker(
    description='Date:',
    disabled=False
)

pickerForecast= widgets.Button(
    layout=widgets.Layout(width='200px'),
    description="Get forecast for picked date")
pickerForecast.on_click(showPickedForecast)

showWeekend= widgets.Button(
    layout=widgets.Layout(width='250px'),
    description="Get forecast for closest weekend")
showWeekend.on_click(getWeekendForecast)

row1 = widgets.HBox([myDate,pickerForecast])

display(row1,showWeekend)

```
